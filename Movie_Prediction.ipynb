```
#visualize some of the data
import pandas as pd
myfile = pd.read_csv('wiki_movie_plots_deduped.csv')
myfile.head()
```

```
#read in the features and labels into lists
import csv
sentences = []
categories = []
with open('wiki_movie_plots_deduped.csv') as csvfile:
  reader = csv.reader(csvfile)
  next(reader)
  for line in reader:
    sentences.append(line[7])
    categories.append(line[2])
print(len(sentences))
print(len(categories))
print(sentences[100])
print(categories[100])
```

```
#only keep these categories: 'American' and 'British' and 'Bollywood'
new_categories = []
new_sents = []
for i in range(len(categories)):
  if categories[i] == 'American' or categories[i] == 'Bollywood' or categories[i] == 'British' :
    new_categories.append(categories[i])
    new_sents.append(sentences[i])
sentences = new_sents
categories = new_categories
print(len(sentences))
print(len(categories))
```

```
#shuffle the data
import random
comb_list = []
for i in range(len(sentences)):
  comb_list.append((sentences[i], categories[i]))
random.shuffle(comb_list)
sentences = []
categories = []
for i in range(len(comb_list)):
  sentences.append(comb_list[i][0])
  categories.append(comb_list[i][1])
print(sentences[0])
print(categories[0])
```

```
#remove stopwords
stopwords = [ "a", "about", "above", "after", "again", "against", "all", "am", "an", "and", "any", "are", "as", "at", "be", "because", "been", "before", "being", "below", "between", "both", "but", "by", "could", "did", "do", "does", "doing", "down", "during", "each", "few", "for", "from", "further", "had", "has", "have", "having", "he", "he'd", "he'll", "he's", "her", "here", "here's", "hers", "herself", "him", "himself", "his", "how", "how's", "i", "i'd", "i'll", "i'm", "i've", "if", "in", "into", "is", "it", "it's", "its", "itself", "let's", "me", "more", "most", "my", "myself", "nor", "of", "on", "once", "only", "or", "other", "ought", "our", "ours", "ourselves", "out", "over", "own", "same", "she", "she'd", "she'll", "she's", "should", "so", "some", "such", "than", "that", "that's", "the", "their", "theirs", "them", "themselves", "then", "there", "there's", "these", "they", "they'd", "they'll", "they're", "they've", "this", "those", "through", "to", "too", "under", "until", "up", "very", "was", "we", "we'd", "we'll", "we're", "we've", "were", "what", "what's", "when", "when's", "where", "where's", "which", "while", "who", "who's", "whom", "why", "why's", "with", "would", "you", "you'd", "you'll", "you're", "you've", "your", "yours", "yourself", "yourselves" ]
for i in range(len(sentences)):
  sentences[i] = sentences[i].lower()
  #also remove (...) type stuff
  for w in stopwords:
    tok = ' ' + w + ' '
    sentences[i] = sentences[i].replace(tok, ' ')
    sentences[i] = sentences[i].replace('  ', ' ')
    
print(sentences[100])
```

```
#split the data into testing and training sets
percent_train = 0.8
train_len = int(percent_train * len(sentences))
train_sentences = sentences[:train_len]
train_labels = categories[:train_len]
test_sentences = sentences[train_len:]
test_labels = categories[train_len:]
print(train_sentences[200], train_labels[200])
print(len(train_sentences), len(train_labels))
print(test_sentences[10], test_labels[10])
print(len(test_sentences), len(test_labels))
```

```
#import the tensorflow machine learning library
#also import other relevant items for preprocessing of the text data
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(oov_token = '<OOV>')
tokenizer.fit_on_texts(train_sentences) #tokenize the words in the train sentences
word_index = tokenizer.word_index
print(word_index)
vocab_size = len(word_index) #store the vocab size for later
print(vocab_size)
my_seqs = tokenizer.texts_to_sequences(train_sentences) #convert the words into numbers
my_seqs = pad_sequences(my_seqs, maxlen = 120, truncating = 'post', padding = 'post') #pad the sequences to make them of equal length
```

```
#follow the same procedure to encode the test sentences 
test_seq = tokenizer.texts_to_sequences(test_sentences)
test_seq = pad_sequences(test_seq, maxlen = 120, truncating='post', padding = 'post')
#check what kind of data is produced
print(test_seq[0])
print(test_sentences[0])
```

```
#follow the same procedure for the labels
label_tokenizer = Tokenizer()
label_tokenizer.fit_on_texts(train_labels)
out_voc_size = len(label_tokenizer.word_index)
print(label_tokenizer.word_index)
out_voc_size += 1
print(out_voc_size)
train_label_seqs = label_tokenizer.texts_to_sequences(train_labels)
train_label_seqs = pad_sequences(train_label_seqs, maxlen = 1)
train_label_seqs = tf.keras.utils.to_categorical(train_label_seqs, num_classes=out_voc_size) #convert the labels into binary arrays
print(len(train_label_seqs[10]))
```

```
test_label_seqs = label_tokenizer.texts_to_sequences(test_labels)
test_label_seqs = pad_sequences(test_label_seqs, maxlen = 1)
test_label_seqs = tf.keras.utils.to_categorical(test_label_seqs, num_classes=out_voc_size)
print(test_label_seqs[150])
print(len(test_label_seqs[150]))
```

```
#define the ML model
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(vocab_size + 1, 64, input_length=120)) #embedding layers (convert the words into vectors)
model.add(tf.keras.layers.Bidirectional(tf.keras.layers.GRU(64, return_sequences=True))) #rnn layer to carry context forward and backward
model.add(tf.keras.layers.Conv1D(64, 5)) #convolutional layer
model.add(tf.keras.layers.GlobalAveragePooling1D()) #pool the results of the previous layer
model.add(tf.keras.layers.Dropout(0.5)) #dropout to reduce overfitting
model.add(tf.keras.layers.Dense(64, activation='relu')) #just a regular dense layer
model.add(tf.keras.layers.Dense(out_voc_size, activation=tf.nn.softmax)) #output layer

model.summary()
```

```
adam = tf.keras.optimizers.Adam(learning_rate=0.001) #define the optimizer
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy']) #compile the model with categorical crossentropy loss
```

```
#train the model
history = model.fit(my_seqs, train_label_seqs, epochs=5, validation_data=(test_seq, test_label_seqs), batch_size=30)
```
